```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set()
```

```python
data = pd.read_excel(r"Data_Train.xlsx")
data.head()
```

```python
data['Duration'].value_counts()
```

```python
# drop Nan NaT
data.dropna(inplace = True)
```

```python
data.isnull().sum()
```

```python
# explotary data analysis
# dt.day method will extract only day of that date
# dt.month  method will extract only month of the date
```

```python
data["Journey_day"]=pd.to_datetime(data.Date_of_Journey, format="%d/%m/%Y").dt.day
```

```python
data["Journey_month"]=pd.to_datetime(data["Date_of_Journey"], format="%d/%m/%Y").dt.month
```

```python
data.head()
```

```python
# since we converted  date of journey in to integer  now we can drop it in use 
data.drop(['Date_of_Journey'],axis=1,inplace=True)
```

```python
#departure time is when the plane leaves the gate
#similar to date of journey we can extract values from deo_time

#Extracting Hours
data["Dep_hour"]=pd.to_datetime(data["Dep_Time"]).dt.hour

#Extracting minutes
data["Dep_Min"]=pd.to_datetime(data["Dep_Time"]).dt.minute

#now we can drop dep_time as it is of no use 
data.drop(["Dep_Time"],axis =1, inplace=True)
```

```python
data.head()
```

```python
# arrival time is when the plane pulls up to the gate 
#similar to date of journey we can extract values from arrival time 

#Extracting Hours
data["Arrival_hour"]=pd.to_datetime(data.Arrival_Time).dt.hour

#Extracting minutes
data["Arrival_min"]=pd.to_datetime(data.Arrival_Time).dt.minute

#now we can drop arrival time as it is of no use 
data.drop(["Arrival_Time"],axis=1, inplace=True)
```

```python
data.head()
```

```python
len
```

```python
#time taken by plane to reach destinating  is called duration 
#it is the difference between departure time and arrival time 
#assinging  and converting duration column into  list 

duration =list(data["Duration"])
for i in range(len(duration)):
    if len(duration[i].split()) !=2:  #check if duration cantains only hours or mins
        if "h" in duration[i]:
            duration[i]=duration[i].strip()+ " 0m" #adds 0 minutes
        else:
            duration[i]="0h " + duration[i]         #adds0 hours
            
duration_hours=[]
duration_mins=[]
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0])) 
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))
```

```python
#adding duration hours and duration mins list to train data data frame
data["Duration_hours"]=duration_hours
data["Duration_mins"]=duration_mins
```

```python
data.head()
```

```python
data.drop(["Duration"],axis=1,inplace=True)
```

```python
data.head()
```

```python
# as  source in nominal categorial  data we will perform oneHotencoding
Source= data[["Source"]]
Source=pd.get_dummies(Source,drop_first= True)
Source.head()
```

```python
data["Destination"].value_counts()
```

```python
# as Destination is nominal categorial data we will perform oneHotEncoding
Destination = data[["Destination"]]
Destination=pd.get_dummies(Destination,drop_first=True)
Destination
```

```python
data["Route"]
```

```python
# Additional info contains almost 80% no info 
# route and total stops are related to each other 
data.drop(["Route","Additional_Info"], axis=1, inplace= True)
```

```python
data.head()
```

```python
data.replace({"non-stop":0,"1 stop":1,"2 stops":2,"3 stops":3,"4 stops":4},inplace=True)
```

```python
data.head()
```

```python
# concatenate dataframe   train data+ Airline+Source+ destination
data=pd.concat([data,Airline,Source,Destination],axis=1)
```
